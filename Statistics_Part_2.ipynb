1.  What is hypothesis testing in statistics

 Hypothesis testing in statistics is a method used to make decisions or inferences about a population based on sample data. It helps you test an assumption (called a hypothesis) about a population parameter, such as the mean or proportion.

Key Concepts of Hypothesis Testing:

Null Hypothesis (H‚ÇÄ):

This is the default or original claim you assume to be true.

Example: The average height of students is 5.5 feet.

Alternative Hypothesis (H‚ÇÅ or Ha):

This is what you want to prove or support.

Example: The average height of students is not 5.5 feet.

Significance Level (Œ±):

It‚Äôs the probability of rejecting the null hypothesis when it is actually true. Common values are 0.05, 0.01, or 0.10.

Test Statistic:

A standardized value calculated from sample data, used to decide whether to reject H‚ÇÄ. Examples include z, t, chi-square.

p-value:

The probability of getting a test statistic as extreme as (or more than) the one observed, assuming H‚ÇÄ is true.

Decision Rule:

If p-value ‚â§ Œ±, reject H‚ÇÄ (evidence supports H‚ÇÅ).

If p-value > Œ±, do not reject H‚ÇÄ (insufficient evidence to support H‚ÇÅ).

Steps in Hypothesis Testing:

State H‚ÇÄ and H‚ÇÅ.

Select significance level (Œ±).

Choose the appropriate test and compute the test statistic.

Find the p-value or critical value.

Make a decision (reject or do not reject H‚ÇÄ).

Interpret the result in context.

Example:

Claim: A company claims that their light bulbs last 1000 hours.

Sample data: A sample of 40 bulbs shows an average life of 980 hours with a known standard deviation of 60 hours.

H‚ÇÄ: Œº = 1000

H‚ÇÅ: Œº ‚â† 1000

Œ± = 0.05

Use a z-test because population standard deviation is known.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

 The null hypothesis and the alternative hypothesis are fundamental concepts in statistics, especially in hypothesis testing.

>1. Null Hypothesis (H‚ÇÄ):

The null hypothesis is a statement that there is no effect, no difference, or no relationship in the population.

It represents the status quo or a baseline assumption that you test against.

Example:

If you're testing whether a new drug is effective, the null hypothesis would be:

H‚ÇÄ: The new drug has no effect (i.e., it is not better than the old one or a placebo).

>2. Alternative Hypothesis (H‚ÇÅ or Ha):

The alternative hypothesis is a statement that contradicts the null hypothesis.

It suggests that there is an effect, a difference, or a relationship.

Example (continued):

H‚ÇÅ: The new drug has an effect (i.e., it is better than the old one or a placebo).

3. What is the significance level in hypothesis testing, and why is it important?

 The significance level in hypothesis testing, often denoted by the Greek letter alpha (Œ±), is the probability of rejecting the null hypothesis (H‚ÇÄ) when it is actually true. In simpler terms, it represents the risk of making a Type I error‚Äîa "false positive."

Common Significance Levels:

0.05 (5%) ‚Äì Most commonly used

0.01 (1%)

0.10 (10%)

Why is Significance Level Important?

Controls Error Rate:

It helps control the probability of making a Type I error, which is crucial in scientific and business decisions.

Sets the Benchmark:

It defines the threshold at which you decide whether to reject or not reject the null hypothesis.

If the p-value < Œ±, we reject H‚ÇÄ.

Determines Critical Value:

The significance level helps determine the critical value(s) which form the rejection region in the distribution curve.

Helps in Decision-Making:

Especially in real-life applications like medical trials, market testing, or quality control, choosing the right Œ± level affects outcomes and risks.

Balances Risk and Confidence:

A lower Œ± reduces the chance of Type I error but increases the chance of Type II error (not rejecting H‚ÇÄ when it's false), and vice versa.

Example:

Imagine testing a new drug.

H‚ÇÄ: The drug has no effect.

H‚ÇÅ: The drug is effective.

Œ± = 0.05 means you are willing to accept a 5% chance of saying the drug works when it actually doesn‚Äôt.

4. What does a P-value represent in hypothesis testing?

 In hypothesis testing, the P-value (or probability value) is a measure that helps you decide whether to reject the null hypothesis (H‚ÇÄ).

What the P-value represents:

It tells you the probability of obtaining a result at least as extreme as the one observed in your sample data, assuming the null hypothesis is true.

In simple terms:

A small P-value (typically ‚â§ 0.05) means the observed result is unlikely under the null hypothesis, so you reject H‚ÇÄ.

A large P-value (typically > 0.05) means the result is likely under the null, so you fail to reject H‚ÇÄ.

Example:

You want to test if a coin is fair (H‚ÇÄ: the coin is fair). You flip it 100 times and get 80 heads.

You calculate the P-value, which might turn out to be 0.001.

This low P-value suggests that getting 80 heads is very unlikely if the coin were truly fair.

So, you reject H‚ÇÄ and conclude the coin is probably biased.

Important Notes:

The P-value does not tell you the probability that the null hypothesis is true.

It only tells you how compatible your data is with the assumption that H‚ÇÄ is true.

5.  How do you interpret the P-value in hypothesis testing?

 The P-value is the probability of obtaining test results at least as extreme as the observed results, assuming that the null hypothesis is true.

 Example:

Suppose you're testing whether a new drug is effective:

H‚ÇÄ: The drug has no effect.

H‚ÇÅ: The drug has a significant effect.

You get a P-value of 0.03.

 Since 0.03 < 0.05, you reject H‚ÇÄ and conclude there's significant evidence the drug works.

 Remember:

A small P-value means the observed data is unlikely under H‚ÇÄ.

A large P-value means the data is more likely under H‚ÇÄ.

P-value does not give the probability that H‚ÇÄ is true.

6.  What are Type 1 and Type 2 errors in hypothesis testing?

 In hypothesis testing, Type 1 and Type 2 errors are two kinds of incorrect decisions you can make when testing a hypothesis. Here's a simple explanation of both:

Type 1 Error (False Positive)

Meaning: Rejecting the null hypothesis when it is actually true.

In simple words: You think there is an effect or difference, but in reality, there isn‚Äôt.

Example: A medical test says a person has a disease (positive result), but the person is actually healthy.

Symbol: Denoted by Œ± (alpha) ‚Äî usually set at 0.05 (5%).

Also called: Error of the first kind.

Type 2 Error (False Negative)

Meaning: Failing to reject the null hypothesis when it is actually false.

In simple words: You think there is no effect or difference, but in reality, there is.

Example: A medical test says a person is healthy (negative result), but the person actually has the disease.

Symbol: Denoted by Œ≤ (beta).

Also called: Error of the second kind.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing

In hypothesis testing, one-tailed and two-tailed tests refer to the direction of the statistical test and the nature of the alternative hypothesis. Here‚Äôs a breakdown of the differences:

 One-Tailed Test (Directional Test)

Definition: A test where the critical region (area of rejection) is only on one side (either left or right) of the sampling distribution.

Used When: You have a specific direction in your alternative hypothesis (either increase or decrease).

Example:

Null Hypothesis (H‚ÇÄ): Œº = 50

Alternative Hypothesis (H‚ÇÅ): Œº > 50 (right-tailed) or Œº < 50 (left-tailed)

Application: Testing if a new drug increases recovery rate compared to the current one.

 Two-Tailed Test (Non-directional Test)

Definition: A test where the critical region is split between both sides of the sampling distribution.

Used When: You are testing for any significant difference (either higher or lower).

Example:

Null Hypothesis (H‚ÇÄ): Œº = 50

Alternative Hypothesis (H‚ÇÅ): Œº ‚â† 50

Application: Testing if a new teaching method results in a different average score (could be higher or lower) than the traditional method.

8. What is the Z-test, and when is it used in hypothesis testing?

 The Z-test is a type of statistical hypothesis test used to determine whether there is a significant difference between sample and population means, or between the means of two samples, when the population variance is known and the sample size is large (usually n ‚â• 30).

 Definition of Z-test:

A Z-test uses the Z-distribution (standard normal distribution) to test hypotheses. The test statistic follows a normal distribution under the null hypothesis.

 When is Z-test used?

You use a Z-test when:

The sample size is large (n ‚â• 30).

The population standard deviation (œÉ) is known.

The data is normally distributed or approximately normal due to the Central Limit Theorem.

You're comparing:

A sample mean with a known population mean, or

Two sample means, or

A sample proportion with a population proportion.

 Types of Z-tests:

One-sample Z-test for mean:

Used to compare the sample mean to a known population mean.

Two-sample Z-test for means:

Compares the means of two independent samples.

Z-test for proportions:

Used to test whether the sample proportion significantly differs from the population proportion.

10.  What is the T-distribution, and when should it be used instead of the normal distribution?

 The T-distribution, also known as Student‚Äôs T-distribution, is a probability distribution that is used in inferential statistics, especially for estimating population parameters when the sample size is small and/or the population standard deviation is unknown.

 What is the T-Distribution?
It is similar in shape to the normal distribution (bell-shaped and symmetric), but it has heavier tails.

As the sample size increases, the T-distribution approaches the normal distribution.

It is defined by degrees of freedom (df), which typically equals n - 1 where n is the sample size.

 When to Use the T-Distribution Instead of the Normal Distribution:
Use the T-distribution when:

 The sample size is small (typically n < 30).

 The population standard deviation (œÉ) is unknown.

 The data is approximately normally distributed or close to normal.

11. What is the difference between a Z-test and a T-test?

 1. Sample Size:

A Z-test is used when your sample size is large, usually more than 30.

A T-test is used when your sample size is small, typically 30 or less.

2. Population Standard Deviation:

In a Z-test, you know the population standard deviation (œÉ).

In a T-test, you don‚Äôt know the population standard deviation, so you use the sample standard deviation (s).

3. Distribution Used:

The Z-test is based on the standard normal distribution (also called the Z-distribution).

The T-test uses the t-distribution, which is wider and more flexible, especially with smaller samples.

4. Accuracy with Small Samples:

The T-test gives more reliable results when the sample size is small because the t-distribution adjusts for uncertainty by using degrees of freedom.

The Z-test assumes the data follows a perfect bell curve, which is more reasonable when the sample is large.

5. When to Use Each:

Use a Z-test if the sample is large and you know the population standard deviation.

Use a T-test if the sample is small and you don‚Äôt know the population standard deviation.

13.  What is the relationship between Z-test and T-test in hypothesis testing?

 The Z-test and the T-test are both tools used in hypothesis testing. Their purpose is similar ‚Äî they help you decide whether to accept or reject a null hypothesis, typically about a population mean or difference between means.

 Similarities:

Both are used to test hypotheses based on sample data.

Both compare a sample statistic (like a sample mean) to a population parameter.

Both calculate a test statistic that is compared to a critical value to make a decision.

 The Main Differences:

Z-test is used when the sample size is large (usually 30 or more) and the population standard deviation is known.

T-test is used when the sample size is small (usually less than 30) and the population standard deviation is unknown. In this case, we use the sample standard deviation instead.

Because we‚Äôre less certain about estimates from small samples, the T-test uses the t-distribution, which is wider and has heavier tails than the standard normal (Z) distribution. This accounts for the increased uncertainty.

14.  What is a confidence interval, and how is it used to interpret statistical results?

 A confidence interval (CI) is a range of values, derived from sample data, that is likely to contain the true population parameter (like a mean or proportion) with a certain level of confidence.

 Definition:

A confidence interval gives an estimated range of values which is likely to include an unknown population parameter.
It is usually expressed as:

Estimate ¬± Margin of Error

 Key Components:

Point Estimate: A single value estimate (e.g., sample mean).

Margin of Error: Reflects the variability in the data and the level of confidence.

Confidence Level: The probability that the interval contains the true parameter. Common levels:

90%

95% (most common)

99%

 Example:

Suppose the average height of a sample of students is 160 cm, and the 95% confidence interval is [158 cm, 162 cm].

 This means we are 95% confident that the true average height of the entire population lies between 158 cm and 162 cm.

 Why It Matters:

It provides more information than just a point estimate.

It shows the reliability and precision of the estimate.

Wider intervals = less precision, Narrower intervals = more precision.

 Important Notes:

A 95% CI means if you repeated the sampling 100 times, about 95 of those intervals would contain the true population parameter.

It does not mean there's a 95% chance that the parameter is in your specific interval (the parameter is fixed, the interval varies).

21. What are the different types of ANOVA tests?

 There are several types of ANOVA (Analysis of Variance) tests, each suited for different types of experimental designs and research questions. Here's a breakdown of the main types of ANOVA tests:

 1. One-Way ANOVA (Single Factor ANOVA)

Purpose: Compares the means of three or more independent (unrelated) groups based on one independent variable (factor).

Example: Comparing average test scores among students from three different schools.

 2. Two-Way ANOVA (Factorial ANOVA)

Purpose: Examines the influence of two independent variables on one dependent variable, including possible interaction effects.

Types:

Without replication: One observation per combination of factors.

With replication: Multiple observations per combination of factors.

Example: Studying the effect of teaching method and gender on exam scores.

 3. Repeated Measures ANOVA

Purpose: Used when the same subjects are measured multiple times under different conditions or time points.

Example: Measuring weight loss of a group after 3 different diets over 3 months.

 4. Mixed-Design ANOVA (Split-Plot ANOVA)

Purpose: Combines between-subjects factors (independent groups) and within-subjects factors (repeated measures).

Example: Analyzing the effect of a new drug (between groups) across multiple weeks (within subjects).

 5. MANOVA (Multivariate Analysis of Variance)

Purpose: Used when there are two or more dependent variables. It tests for differences in multiple continuous dependent variables based on one or more factors.

Example: Investigating how education level affects both income and job satisfaction.

 6. ANCOVA (Analysis of Covariance)

Purpose: Combines ANOVA and regression. It tests group differences while controlling for the effects of covariates (continuous variables).

Example: Comparing the effectiveness of different therapies on depression levels while controlling for age.

```python
## PRACTICLA QUESTIONS

#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
 #interpret the results.

import numpy as np
from scipy import stats

# Sample data
sample_data = [105, 110, 102, 108, 115, 98, 107, 111, 100, 109]

# Known population parameters
population_mean = 100
population_std = 10  # Known population standard deviation

# Calculate sample statistics
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

# Calculate the Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate the p-value (two-tailed)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Print the results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the result
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

```python
#9.  Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import scipy.stats as stats
import numpy as np

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
        data (list or array): Sample data
        confidence (float): Confidence level (default is 0.95 for 95%)

    Returns:
        (lower_bound, upper_bound): Tuple representing the confidence interval
    """
    sample_mean = np.mean(data)
    sample_size = len(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation (ddof=1)

    # Calculate the standard error
    standard_error = sample_std / np.sqrt(sample_size)

    # Find the t-score for the given confidence level and degrees of freedom
    t_score = stats.t.ppf((1 + confidence) / 2.0, df=sample_size - 1)

    # Calculate the margin of error
    margin_of_error = t_score * standard_error

    # Return the confidence interval
    return (sample_mean - margin_of_error, sample_mean + margin_of_error)

data = [10, 12, 13, 15, 16, 18, 20]
ci = confidence_interval(data, confidence=0.95)
print("95% Confidence Interval:", ci)

```

```python
#11.  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
# Given probabilities
P_D = 0.01           # Prior: Probability of disease
P_not_D = 1 - P_D    # Probability of no disease

P_Pos_given_D = 0.99     # Sensitivity: P(Test+ | Disease)
P_Pos_given_not_D = 0.05 # 1 - Specificity: P(Test+ | No Disease)

# Bayes' Theorem:
# P(Disease | Test+) = [P(Test+ | Disease) * P(Disease)] / P(Test+)

# First, calculate the marginal probability of a positive test
P_Pos = (P_Pos_given_D * P_D) + (P_Pos_given_not_D * P_not_D)

# Now calculate the posterior
P_D_given_Pos = (P_Pos_given_D * P_D) / P_Pos

print(f"Probability of having the disease given a positive test: {P_D_given_Pos:.4f}")
```

```python
#13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
#data
def calculate_expected_frequencies(observed):
    rows = len(observed)
    cols = len(observed[0])

    # Calculate row totals and column totals
    row_totals = [sum(row) for row in observed]
    col_totals = [sum(observed[row][col] for row in range(rows)) for col in range(cols)]
    grand_total = sum(row_totals)

    # Calculate expected frequencies
    expected = [[(row_totals[i] * col_totals[j]) / grand_total for j in range(cols)] for i in range(rows)]

    return expected

# Example usage:
observed_data = [
    [90, 60, 104, 95],
    [30, 50, 51, 20],
    [30, 40, 45, 35]
]

expected_frequencies = calculate_expected_frequencies(observed_data)

print("Expected Frequencies:")
for row in expected_frequencies:
    print(row)
```

```python
#16. Implement an F-test using Python to compare the variances of two random samples
import numpy as np
from scipy.stats import f

# Sample data
sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=55, scale=15, size=30)

# Calculate sample variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# Calculate F-statistic
F = var1 / var2 if var1 > var2 else var2 / var1

# Degrees of freedom
df1 = len(sample1) - 1
df2 = len(sample2) - 1

# Calculate p-value
p_value = 1 - f.cdf(F, df1, df2)

print(f"F-statistic: {F:.4f}")
print(f"P-value: {p_value:.4f}")
```

```python
#19.  Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA
import pandas as pd
import numpy as np
from scipy.stats import shapiro, levene
from statsmodels.stats.stattools import durbin_watson
from statsmodels.formula.api import ols
import statsmodels.api as sm

def check_anova_assumptions(df, dv, iv):
    """
    Checks ANOVA assumptions: normality, independence, and equal variance.

    Parameters:
    - df: pandas DataFrame
    - dv: dependent variable (string)
    - iv: independent variable (string)
    """
    print("Checking ANOVA Assumptions:\n")

    # Fit the ANOVA model
    model = ols(f"{dv} ~ C({iv})", data=df).fit()
    residuals = model.resid

    # 1. Normality of residuals
    stat, p = shapiro(residuals)
    print("1. Normality (Shapiro-Wilk Test):")
    print(f"   Statistic = {stat:.4f}, p-value = {p:.4f}")
    print("   -> " + ("Assumption met ‚úÖ\n" if p > 0.05 else "Assumption violated ‚ùå\n"))

    # 2. Independence of residuals
    dw_stat = durbin_watson(residuals)
    print("2. Independence (Durbin-Watson):")
    print(f"   Durbin-Watson = {dw_stat:.4f}")
    print("   -> " + ("Assumption met ‚úÖ\n" if 1.5 < dw_stat < 2.5 else "Assumption violated ‚ùå\n"))

    # 3. Homogeneity of variance
    groups = [group[dv].values for name, group in df.groupby(iv)]
    stat, p = levene(*groups)
    print("3. Equal Variance (Levene‚Äôs Test):")
    print(f"   Statistic = {stat:.4f}, p-value = {p:.4f}")
    print("   -> " + ("Assumption met ‚úÖ\n" if p > 0.05 else "Assumption violated ‚ùå\n"))

# Sample data to test
data = {
    'score': [23, 25, 26, 28, 29, 27, 30, 32, 31],
    'group': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C']
}
df = pd.DataFrame(data)

# Run the assumption check
check_anova_assumptions(df, dv='score', iv='group')
```

```python
#21.  Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Set degrees of freedom for numerator and denominator
dfn = 5  # degrees of freedom numerator
dfd = 10  # degrees of freedom denominator

# Generate x values
x = np.linspace(0.01, 5, 500)

# F-distribution PDF
y = f.pdf(x, dfn, dfd)

# Plot the F-distribution
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', label=f'F-distribution (dfn={dfn}, dfd={dfd})')
plt.fill_between(x, y, where=(x > f.ppf(0.95, dfn, dfd)), color='red', alpha=0.3, label='Critical Region (5%)')

plt.title('F-distribution PDF')
plt.xlabel('F value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

# Print critical value at 5% significance level
critical_value = f.ppf(0.95, dfn, dfd)
print(f"Critical value at 5% significance level: {critical_value:.4f}")

# Explanation:
print("\nUse in Hypothesis Testing:")
print("The F-distribution is used in ANOVA and variance ratio tests.")
print("It helps test if the variances of two populations are equal.")
print("If the test statistic exceeds the critical value, we reject the null hypothesis.")
```

```python
#23.  Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means
import numpy as np
from scipy import stats

# Step 1: Simulate random data from a normal distribution
np.random.seed(42)  # for reproducibility
data = np.random.normal(loc=50, scale=10, size=100)  # mean=50, std=10, n=100

# Step 2: Perform a one-sample t-test to test if the mean is equal to 52
t_statistic, p_value = stats.ttest_1samp(data, popmean=52)

# Output the results
print(f"Sample Mean: {np.mean(data):.2f}")
print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Reject the null hypothesis: The mean is significantly different from 52.")
else:
    print("Fail to reject the null hypothesis: The mean is not significantly different from 52.")
```

```python
#24.  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results
import scipy.stats as stats
import numpy as np

# Sample data
data = [12, 15, 14, 10, 13, 15, 14, 13, 11, 12]  # Example sample
sample_size = len(data)
sample_variance = np.var(data, ddof=1)  # Sample variance (unbiased)
hypothesized_variance = 4  # H0: œÉ¬≤ = 4

# Chi-square test statistic
chi_square_stat = (sample_size - 1) * sample_variance / hypothesized_variance

# Degrees of freedom
df = sample_size - 1

# Significance level
alpha = 0.05

# Critical values
chi2_lower = stats.chi2.ppf(alpha / 2, df)
chi2_upper = stats.chi2.ppf(1 - alpha / 2, df)

# Interpretation
if chi2_lower < chi_square_stat < chi2_upper:
    result = "Fail to reject H0: There is not enough evidence to say population variance is different from hypothesized."
else:
    result = "Reject H0: The population variance is significantly different from hypothesized."

# Output
print(f"Chi-square Statistic = {chi_square_stat:.2f}")
print(f"Critical Values = ({chi2_lower:.2f}, {chi2_upper:.2f})")
print("Result:", result)
```

```python
#26.  Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Sample datasets
data1 = np.random.normal(loc=50, scale=10, size=100)
data2 = np.random.normal(loc=55, scale=15, size=100)

# Calculate variances
var1, var2 = np.var(data1, ddof=1), np.var(data2, ddof=1)

# F-test statistic
f_stat = var1 / var2 if var1 > var2 else var2 / var1
df1, df2 = len(data1) - 1, len(data2) - 1
p_value = 1 - stats.f.cdf(f_stat, df1, df2)

# Interpretation
alpha = 0.05
result = "Significant difference in variances" if p_value < alpha else "No significant difference in variances"

# Print results
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Conclusion: {result}")

# Visualization
plt.boxplot([data1, data2], labels=['Dataset 1', 'Dataset 2'])
plt.title('Comparison of Variances using Boxplot')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```
